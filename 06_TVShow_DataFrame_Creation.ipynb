## **Creating TV Show DataFrame**

```python
# Verify that all scraped TV show data lists have the same number of elements.
# This is an important validation step before creating the DataFrame.
# Each list should contain one value for every TV show scraped.

print("Total TV Show Titles:", len(tv_show_titles))
print("Total Release Years:", len(release_years))
print("Total Genres:", len(genres))
print("Total TV Show URLs:", len(tv_show_urls))
print("Total IMDb Ratings:", len(imdb_ratings))
print("Total Runtimes:", len(durations))
print("Total Age Ratings:", len(age_ratings))
print("Total Production Countries:", len(production_countries))
print("Total OTT Streaming Services:", len(primary_ott))
```

```python
# Import the Pandas library
# Pandas is used to organize and analyze structured data
# in the form of DataFrames.
import pandas as pd

# Create a DataFrame by combining all the scraped TV show details.
# Each key in the dictionary represents a column name,
# and the corresponding list contains the data for that column.
tv_shows_df = pd.DataFrame({

    # List of TV show titles extracted from the TV show pages
    "TV_Show_Title": tv_show_titles,

    # List of release years for each TV show
    "Release_Year": release_years,

    # List of genres associated with each TV show
    "Genres": genres,

    # List of URLs corresponding to each TV show page
    "URLs": tv_show_urls,

    # List of IMDb ratings for each TV show
    "IMDb_Rating": imdb_ratings,

    # List of runtimes (episode durations) for each TV show
    "Runtime": durations,

    # List of age ratings (e.g., UA16+, A, PG-13)
    "Age_Rating": age_ratings,

    # List of production countries for each TV show
    "Production_Country": production_countries,

    # List of primary OTT streaming platforms
    # where each TV show is available
    "OTT_Streaming_Service": primary_ott
})

# Display the complete DataFrame.
# Each row represents one TV show and contains all the
# extracted information associated with that TV show.

tv_shows_df
```

```python
# Display a concise summary of the TV Shows DataFrame.
# The info() method provides useful information about the dataset,
# including the number of rows, column names, data types,
# non-null values, and memory usage.

tv_shows_df.info()
```

```python
# Save the TV Shows DataFrame as a CSV file.
# The file will be created in the current working directory.
# Setting index=False prevents Pandas from writing the row index
# as an extra column in the CSV file.

tv_shows_df.to_csv(
    "TV_Shows_Data.csv",
    index=False
)

# Display a confirmation message after the CSV file
# has been successfully saved.

print("TV_Shows_Data.csv saved successfully")
```

```python
# Convert the 'IMDb_Rating' column from text (string) to numeric format.
# The parameter errors="coerce" converts any invalid or missing values
# into NaN (Not a Number), preventing conversion errors.
tv_shows_df["IMDb_Rating"] = pd.to_numeric(
    tv_shows_df["IMDb_Rating"],
    errors="coerce"
)

# Filter the DataFrame to include only TV shows
# with an IMDb rating of 7 or higher.
tv_filtered_df = tv_shows_df[
    tv_shows_df["IMDb_Rating"] >= 7
]

# Display the filtered DataFrame.
# The output will contain only TV shows that satisfy
# the IMDb rating criterion.
tv_filtered_df
```

```python
# Save the filtered TV Shows DataFrame as a CSV file.
# The file will contain only TV shows with an IMDb rating of 7 or higher.
# Setting index=False prevents Pandas from writing the row index
# as an additional column in the CSV file.

tv_filtered_df.to_csv(
    "TV_Filtered_Data.csv",
    index=False
)

# Display a confirmation message after the CSV file
# has been successfully saved.

print("TV_Filtered_Data.csv saved successfully")
```

```python
# Display a concise summary of the filtered TV Shows DataFrame.
# The info() method provides useful details about the filtered dataset,
# including the number of rows, column names, data types,
# non-null values, and memory usage.

tv_filtered_df.info()
```
