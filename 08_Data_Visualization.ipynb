```python
# Word Cloud for Movies Genres

from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Combine all movie genres into a single string
movie_genre_text = " ".join(
    movies_df["Genres"]
    .dropna()
    .str.replace(",", " ", regex=False)
    .tolist()
)

# Generate Word Cloud
movie_wordcloud = WordCloud(
    width=1000,
    height=500,
    background_color="white"
).generate(movie_genre_text)

# Display
plt.figure(figsize=(14, 7))
plt.imshow(movie_wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Most Frequent Movie Genres")
plt.show()
```
