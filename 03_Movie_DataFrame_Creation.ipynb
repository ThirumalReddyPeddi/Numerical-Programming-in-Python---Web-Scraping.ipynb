## **Now Creating Movies DataFrame**

```python
# Verify that all scraped data lists have the same number of elements
# This is an important validation step before creating the DataFrame.
# Each list should have one entry for every movie scraped.

print("Total Movie Titles:", len(movie_titles))
print("Total Release Years:", len(release_years))
print("Total Movie Genres:", len(movie_genres))
print("Total IMDb Ratings:", len(imdb_ratings))
print("Total Movie Runtimes:", len(movie_runtime))
print("Total Age Ratings:", len(age_ratings))
print("Total Production Countries:", len(production_countries))
print("Total Primary OTT Platforms:", len(primary_ott))
print("Total Movie URLs:", len(movie_urls))
```

```python
# Import the Pandas library
# Pandas is used to create and manipulate tabular data structures called DataFrames
import pandas as pd

# Create a DataFrame by combining all the scraped movie information
# Each key in the dictionary represents a column name,
# and the corresponding list contains the values for that column
movies_df = pd.DataFrame({

    # List of movie titles extracted from the movie pages
    "Movie_Title": movie_titles,

    # List of movie release years
    "Release_Year": release_years,

    # List of movie genres
    "Genres": movie_genres,

    # List of URLs corresponding to each movie page
    "URLs": movie_urls,

    # List of IMDb ratings for each movie
    "IMDb_Rating": imdb_ratings,

    # List of movie runtimes
    "Runtime": movie_runtime,

    # List of age ratings (e.g., U, U/A, A, PG-13)
    "Age_Rating": age_ratings,

    # List of production countries for each movie
    "Production_Country": production_countries,

    # List of primary OTT streaming platforms
    # where each movie is available
    "OTT Streaming_Service": primary_ott
})

# Display the first five rows of the DataFrame
# This helps verify that the scraped data has been correctly stored
movies_df.head()
```

```python
import pandas as pd

movies_df = pd.DataFrame({
    "Movie_Title": movie_titles,
    "Release_Year": release_years,
    "Genres": movie_genres,
    'URLs':movie_urls,
    "IMDb_Rating": imdb_ratings,
    "Runtime": movie_runtime,
    "Age_Rating": age_ratings,
    "Production_Country": production_countries,
    "OTT Streaming_Service": primary_ott
})

# Display the complete DataFrame
# Each row represents one movie and its associated details
movies_df
```

```python
# Display a concise summary of the DataFrame
# The info() method provides important information about the dataset,
# including the number of rows, column names, data types,
# non-null values, and memory usage.

movies_df.info()
```

```python
# Save the movies DataFrame as a CSV file
# The file will be created in the current working directory.
# Setting index=False prevents Pandas from writing the row index
# as an additional column in the CSV file.

movies_df.to_csv(
    "Movies_Dataa.csv",
    index=False
)

# Display a confirmation message after the file is successfully saved
print("Movies_Data.csv saved successfully")
```

```python
# Convert the 'IMDb_Rating' column from text (string) to numeric format
# The 'errors="coerce"' parameter replaces any invalid or missing values
# with NaN (Not a Number), preventing conversion errors.

movies_df["IMDb_Rating"] = pd.to_numeric(
    movies_df["IMDb_Rating"],
    errors="coerce"
)

# Filter the DataFrame to include only movies
# with an IMDb rating of 7 or higher.

filtered_df = movies_df[
    movies_df["IMDb_Rating"] >= 7
]

# Display the filtered DataFrame
# The output will contain only movies that satisfy the IMDb rating condition.

filtered_df
```

```python
# Save the filtered DataFrame as a CSV file
# The file will contain only movies with an IMDb rating of 7 or higher.
# Setting index=False prevents the DataFrame row index
# from being written as an extra column in the CSV file.

filtered_df.to_csv(
    "Filtered_Movies_Data.csv",
    index=False
)

# Display a confirmation message after the CSV file is successfully saved

print("Filtered_Movies_Data.csv saved successfully")
```
