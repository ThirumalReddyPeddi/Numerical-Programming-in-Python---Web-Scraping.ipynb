## **Scrapping Movies Data**

````python
# Function to send an HTTP request to the given URL
# and return the parsed HTML content as a BeautifulSoup object.

def fetch_movie_urls(url):

    # Define request headers to mimic a real web browser.
    # This helps prevent the website from blocking the request.

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
    }

    # Send an HTTP GET request to the target URL.

    response = requests.get(url, headers=headers)


    # Check whether the request was successful.
    # A status code other than 200 indicates that the page
    # could not be retrieved properly.

    if response.status_code != 200:
        return "Failed to retrieve the page, status code:", response.status_code

    # Parse the HTML content of the page using BeautifulSoup.

    soup = BeautifulSoup(response.text, 'html.parser')

    # Return the parsed HTML object for further data extraction.

    return soup


# URL of the JustWatch movies page.
# The parameter 'release_year_from=2000' filters movies released from the year 2000 onwards.

url = 'https://www.justwatch.com/in/movies?release_year_from=2000'

# Call the function to fetch and parse the webpage.
soup=fetch_movie_urls(url)


# Print the formatted HTML content of the webpage.
# This is useful for inspecting the page structure
# and identifying the tags/classes needed for web scraping.

print(soup.prettify())
````

## **Fetching Movie URL's**

````python
# Write Your Code here
````

````python
import requests
from bs4 import BeautifulSoup

url = "https://www.justwatch.com/in/movies?release_year_from=2000"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text, "html.parser")

# Create an empty list to store the movie URLs
movie_urls = []


# Find all anchor (<a>) tags that contain an href attribute
for a in soup.find_all("a", href=True):
   # Extract the href (link) from the current anchor tag
    href = a["href"]


    # Check whether the link corresponds to a movie page
    # Movie URLs on JustWatch contain '/movie/'
    if "/movie/" in href:

        # Convert the relative URL into a complete (absolute) URL
        full_url = "https://www.justwatch.com" + href

        # Add the full movie URL to the list
        movie_urls.append(full_url)

# Remove duplicate movie URLs by converting the list to a set and back to a list
movie_urls = list(set(movie_urls))

# Display the total number of unique movie URLs extracted
print(len(movie_urls))

# Create a final list containing all the extracted movie URLs
# (This step is optional since movie_urls already contains the required data)
final_url = [url for url in movie_urls]

# Print the list of all extracted movie URLs
print(final_url)
````

## **Scrapping Movie Title**

````python
import requests
from bs4 import BeautifulSoup

# Create an empty list to store the titles of all movies
movie_titles = []


# Loop through each movie URL extracted from the JustWatch website
for movie_url in movie_urls:


    # Send an HTTP GET request to the current movie URL
    response = requests.get(
        movie_url,
        headers=headers
    )

# Parse the HTML content of the movie page
    soup = BeautifulSoup(
        response.text,
        "html.parser"
    )

    # Locate the <h1> tag, which contains the movie title
    title_tag = soup.find("h1")

# Check whether the title tag exists
    if title_tag:

        # Extract the text from the <h1> tag, remove any extra spaces,
        # and append the movie title to the list
        movie_titles.append(
            title_tag.get_text(strip=True)
        )
    else:

        # If no title is found, append None as a placeholder
        movie_titles.append(None)


# Print the list of all extracted movie titles
print(movie_titles)

# Print the total number of movie titles extracted
print(len(movie_titles))
````

## **Scrapping Genres**

````python
# Write Your Code here
````

````python

# Create an empty list to store the genres of all movies
movie_genres = []

# Loop through each movie URL extracted from the JustWatch website
for movie_url in movie_urls:

    # Send an HTTP GET request to the current movie page
    response = requests.get(
        movie_url,
        headers=headers
    )

    # Parse the HTML content of the movie page
    soup = BeautifulSoup(
        response.text,
        "html.parser"
    )

    # Find the <h3> tag whose text is "Genres"
    # This acts as the heading for the movie genre section
    genre_heading = soup.find(
        "h3",
        class_="poster-detail-infos__subheading",
        string="Genres"
    )

    # Check whether the Genres section exists
    if genre_heading:

        # Find the next <div> containing the genre information
        genre_div = genre_heading.find_next(
            "div",
            class_="poster-detail-infos__value"
        )

        # Extract the genre text, remove extra spaces,
        # and clean the formatting
        genres = genre_div.get_text(
            separator=" ",
            strip=True
        ).replace(" , ", ", ")

        # Add the extracted genres to the list
        movie_genres.append(genres)

    else:
        # If no genre information is found, store None
        movie_genres.append(None)

# Print the total number of genre entries extracted
print("Total Movie Genres Scraped:", len(movie_genres))

# Print the genres of each movie
print("Movie Genres:")
for movie_genre in movie_genres:
    print(movie_genre)
````

````python
# Create an empty list to store the IMDb ratings of all movies
imdb_ratings = []

for movie_url in movie_urls:

    response = requests.get(movie_url, headers=headers)

    soup = BeautifulSoup(response.text, "html.parser")

    # Find the IMDb logo image (<img> tag with alt="IMDB")
    # This helps locate the IMDb rating section on the page

    imdb_tag = soup.find("img", alt="IMDB")

    # Check whether the IMDb section exists
    if imdb_tag:

        # Find the next <div> element that contains the IMDb rating            rating_div = imdb_tag.find_next("div")

        rating_div = imdb_tag.find_next("div")

        # Extract the IMDb rating text, remove extra spaces,
        # and keep only the rating value
        # Example: "7.8 (171k)" -> "7.8"

        imdb_rating = rating_div.get_text(strip=True).split()[0]

        # Add the extracted IMDb rating to the list
        imdb_ratings.append(imdb_rating)

    else:
        # If the IMDb rating is not available, store None
        imdb_ratings.append(None)

# Print the list of all extracted IMDb ratings
print(imdb_ratings)


 # Print the total number of IMDb ratings extracted
print(len(imdb_ratings))
````

````python
# Create an empty list to store the production country of all movies
production_countries = []

for movie_url in movie_urls:

    response = requests.get(
        movie_url,
        headers=headers
    )

    soup = BeautifulSoup(
        response.text,
        "html.parser"
    )

    # Find the <h3> tag whose text is "Production country"
    # This acts as the heading for the production country section
    country_heading = soup.find(
        "h3",
        class_="poster-detail-infos__subheading",
        string="Production country"
    )

    # Check whether the Production Country section exists
    if country_heading:
        # Find the next <div> that contains the production country information
        country_div = country_heading.find_next(
            "div",
            class_="poster-detail-infos__value"
        )

        # Extract the production country text
        # (e.g., "United States", "India", "United Kingdom")
        # and add it to the production_countries list
        production_countries.append(
            country_div.get_text(strip=True)
        )

    else:
      # If the production country is not available, store None
        production_countries.append(None)

# Print the total number of production country entries extracted
print(len(production_countries))


# Print the production country of each movie
for production_country in production_countries:
    print(production_country)
````

## **Fetching Streaming Service Details**

````python
# Write Your Code here
````

````python
# Create an empty list to store the streaming services available for each movie
streaming_services = []

# Loop through each movie URL extracted from the JustWatch website
for movie_url in movie_urls:

    # Send an HTTP GET request to the current movie page
    response = requests.get(
        movie_url,
        headers=headers
    )

    # Parse the HTML content of the movie page
    soup = BeautifulSoup(
        response.text,
        "html.parser"
    )

    # Find all provider icons (<img> tags) that represent
    # the streaming platforms where the movie is available
    provider_imgs = soup.find_all(
        "img",
        class_="provider-icon"
    )

    # Extract the value of the 'alt' attribute from each image
    # The alt attribute contains the name of the streaming service
    providers = [
        img.get("alt")
        for img in provider_imgs
    ]

    # Combine all provider names into a single comma-separated string
    # and store it in the streaming_services list
    streaming_services.append(
        ", ".join(providers)
    )

# Define a set of valid OTT/streaming platforms
# Only these platforms will be considered while selecting the primary OTT service
valid_services = {
    "Netflix",
    "Amazon Prime Video",
    "Amazon Prime Video with Ads",
    "JioHotstar",
    "Zee5",
    "Sun Nxt",
    "Sony LIV",
    "Aha",
    "Apple TV Store",
    "Amazon Video",
    "Hoichoi",
    "Eros Now",
    "MX Player",
    "Plex Player",
    "JustWatch",
    "Crunchyroll",
}

# Create an empty list to store the primary OTT platform for each movie
primary_ott = []

# Loop through the extracted streaming service strings
for services in streaming_services:

    # Split the comma-separated string into a list of individual services
    service_list = [
        s.strip()
        for s in services.split(",")
    ]

    # Find the first service that matches one of the valid OTT platforms
    # If no valid service is found, return None
    ott = next(
        (
            s
            for s in service_list
            if s in valid_services
        ),
        None
    )

    # Store the selected primary OTT platform
    primary_ott.append(ott)

# Print the total number of primary OTT entries extracted
print("Total Primary OTT Platforms Identified:", len(primary_ott))

# Print the primary OTT platform for each movie
print("Primary OTT Platform for Each Movie:")
for main_ott in primary_ott:
    print(main_ott)
````

````python
# Create an empty list to store the titles of all TV shows
tv_show_titles = []

# Loop through each TV show URL extracted from the JustWatch website
for url in tv_show_urls:

    try:
        # Send an HTTP GET request to the current TV show page
        response = requests.get(
            url,
            headers=headers
        )

        # Parse the HTML content of the TV show page
        soup = BeautifulSoup(
            response.text,
            "html.parser"
        )

        # Find the <h1> tag that contains the TV show title
        # The title is located inside the element with the class
        # 'title-detail-hero__details__title'
        title_tag = soup.select_one(
            "h1.title-detail-hero__details__title"
        )

        # Check whether the title element exists
        if title_tag:

            # Extract the TV show title, remove any extra spaces,
            # and store it in the list
            title = title_tag.contents[0].strip()

        else:
            # If the title is not available, store NaN
            title = np.nan

        # Add the extracted title to the tv_show_titles list
        tv_show_titles.append(title)

    except Exception as e:
        # If any error occurs while fetching or parsing the page,
        # store NaN to maintain alignment with the URL list
        tv_show_titles.append(np.nan)

# Print the total number of TV show titles extracted
print("Total TV Show Titles Scraped:", len(tv_show_titles))

# Print the list of extracted TV show titles
print("TV Show Titles:")
print(tv_show_titles)
````

## **Fetching TV Show Genre Details**

````python
# Create an empty list to store the genres of all TV shows
genres = []

for url in tv_show_urls:


    response = requests.get(
        url,
        headers=headers
    )


    soup = BeautifulSoup(
        response.text,
        "html.parser"
    )

    # Initialize the genre variable with NaN
    # This value will be used if no genre information is found

    genre = np.nan

    # Find all <h3> elements that represent information headings
    # such as Genres, Age rating, Runtime, etc.

    for heading in soup.find_all(
        "h3",
        class_="poster-detail-infos__subheading"
    ):

        # Check whether the current heading is "Genres"

        if heading.get_text(strip=True) == "Genres":

            # Find the next <div> element containing the genre details

            genre_div = heading.find_next(
                "div",
                class_="poster-detail-infos__value"
            )

            # Extract the genre text, remove extra spaces,
            # and clean the formatting
            # Example:
            # "Drama, Crime , Organized Crime"
            # becomes
            # "Drama, Crime, Organized Crime"

            genre = genre_div.get_text(
                separator=" ",
                strip=True
            ).replace(" , ", ", ")

            # Stop searching once the genre section is found

            break

    # Add the extracted genre information to the list

    genres.append(genre)

# Print the total number of genre entries extracted
print( len(genres))

# Print the list of extracted TV show genres

print(genres)
````

## **Fetching IMDB Rating Details**

````python
# Write Your Code here
````

````python
# Create an empty list to store the IMDb ratings of all TV shows
imdb_ratings = []

for url in tv_show_urls:

    try:

        response = requests.get(
            url,
            headers=headers
        )

        soup = BeautifulSoup(
            response.text,
            "html.parser"
        )

        # Find the HTML element that contains the IMDb rating
        # The IMDb score is stored inside the <span> tag
        # with the class "imdb-score"

        imdb_tag = soup.select_one(
            "span.imdb-score"
        )

        # Check whether the IMDb rating element exists
        # If found, extract only the rating value.
        # Example:
        # "8.4 (1.3k)" -> "8.4"
        # Otherwise, store NaN.

        imdb_rating = (
            imdb_tag.get_text(strip=True).split()[0]
            if imdb_tag else np.nan
        )

        # Add the extracted IMDb rating to the list
        imdb_ratings.append(imdb_rating)

    except Exception:
        # If any error occurs while fetching or parsing the page,
        # store NaN to maintain alignment with the URL list.

        imdb_ratings.append(np.nan)

# Print the total number of IMDb ratings extracted
print(len(imdb_ratings))

# Print the list of extracted IMDb ratings

print(imdb_ratings)
````

````python
# Create an empty list to store the age ratings of all TV shows

age_ratings = []


for url in tv_show_urls:


    response = requests.get(
        url,
        headers=headers
    )


    soup = BeautifulSoup(
        response.text,
        "html.parser"
    )

    # Find the <h3> tag whose text is "Age rating"
    # This heading identifies the section containing the age rating information
    age_heading = soup.find(
        "h3",
        string="Age rating"
    )

    # Check whether the Age Rating section exists
    if age_heading:

        # Find the next <div> element containing the age rating value
        # and extract its text

        age_rating = age_heading.find_next(
            "div",
            class_="poster-detail-infos__value"
        ).get_text(strip=True)

    else:
        # If the age rating is not available, store NaN
        age_rating = np.nan

    # Add the extracted age rating to the list
    age_ratings.append(age_rating)

# Print the total number of age ratings extracted
print(len(age_ratings))

# Print the list of extracted age ratings

print(age_ratings)
````

````python
# Create an empty list to store the production country of all TV shows
production_countries = []


for url in tv_show_urls:

    response = requests.get(
        url,
        headers=headers
    )


    soup = BeautifulSoup(
        response.text,
        "html.parser"
    )

    # Find the <h3> tag whose text is "Production country"
    # This heading identifies the section containing
    # the production country information.

    country_heading = soup.find(
        "h3",
        string="Production country"
    )

    # Check whether the Production Country section exists
    # If found, locate the next <div> containing the country name
    # and extract the text. Otherwise, store NaN.

    production_countries.append(
        country_heading.find_next(
            "div",
            class_="poster-detail-infos__value"
        ).get_text(strip=True)
        if country_heading else np.nan
    )

# Print the total number of production country entries extracted
print(len(production_countries))

# Print the list of extracted production countries

print(production_countries)
````

## **Fetching Streaming Service details**

````python
# Write Your Code here
````

````python
# Create an empty list to store the streaming services
# available for each TV show
streaming_services = []

for url in tv_show_urls:


    response = requests.get(
        url,
        headers=headers
    )


    soup = BeautifulSoup(
        response.text,
        "html.parser"
    )

    # Find all provider icons (<img> tags) that represent
    # the streaming platforms where the TV show is available

    providers = soup.find_all(
        "img",
        class_="provider-icon"
    )

    # Check whether any streaming providers are available
    if providers:

        # Extract the streaming service name from the
        # 'title' attribute of each provider icon

        service_names = [
            provider.get("title")
            for provider in providers
            if provider.get("title")
        ]

        # Remove duplicate service names while preserving
        # their original order and combine them into a
        # comma-separated string

        streaming_service = ", ".join(
            dict.fromkeys(service_names)
        )

    else:
        # If no streaming provider is found, store NaN
        streaming_service = np.nan

    # Add the extracted streaming service information
    # to the streaming_services list

    streaming_services.append(streaming_service)


# Define the set of OTT platforms that will be considered
# as valid primary streaming services

valid_services = {
    "Netflix",
    "Amazon Prime Video",
    "Amazon Prime Video with Ads",
    "JioHotstar",
    "Zee5",
    "Sun Nxt",
    "Sony LIV",
    "Aha",
    "Apple TV Store",
    "Amazon Video",
    "Hoichoi",
    "Eros Now",
    "MX Player",
    "Plex Player",
    "JustWatch",
    "Crunchyroll"
}

# Create an empty list to store the primary OTT platform
# for each TV show
primary_ott = []

# Loop through the extracted streaming service strings
for services in streaming_services:

    # Split the comma-separated services into a list
    # and remove any extra spaces
    service_list = [
        s.strip()
        for s in services.split(",")
    ]

    # Select the first service that belongs to the
    # predefined valid OTT platform list
    # If no valid service is found, return None
    ott = next(
        (
            s
            for s in service_list
            if s in valid_services
        ),
        None
    )

    # Add the selected primary OTT platform
    # to the primary_ott list
    primary_ott.append(ott)

# Print the total number of primary OTT platform entries
print(len(primary_ott))

# Print the primary OTT platform for each TV show

for prim_ott in primary_ott:
    print(prim_ott)
````

## **Fetching Duration Details**

````python
# Write Your Code here
````

````python
# Create an empty list to store the duration (runtime)
# of all TV shows
durations = []


for url in tv_show_urls:


    response = requests.get(
        url,
        headers=headers
    )


    soup = BeautifulSoup(
        response.text,
        "html.parser"
    )

    # Initialize the duration variable with NaN
    # This value will be used if no duration is found

    duration = np.nan

    # Find all <div> elements containing title details
    # such as JustWatch rank, IMDb score, and duration

    for item in soup.find_all(
        "div",
        class_="title-detail-hero-details__item"
    ):

        # Extract the text from the current detail item
        text = item.get_text(strip=True)

        # Check whether the text represents the TV show duration
        # Duration values usually contain "min"
        # Examples: "50min", "51min", "1h 6min"

        if "min" in text:

            # Store the extracted duration
            duration = text

            # Stop searching once the duration is found
            break

    # Add the extracted duration to the durations list
    durations.append(duration)

# Print the total number of duration entries extracted
print(len(durations))

# Print the duration of each TV show

for runtime in durations:
    print(runtime)
````
