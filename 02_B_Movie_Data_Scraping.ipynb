##**Fetching Movie URL's**

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.justwatch.com/in/movies?release_year_from=2000"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text, "html.parser")

# Create an empty list to store the movie URLs
movie_urls = []


# Find all anchor (<a>) tags that contain an href attribute
for a in soup.find_all("a", href=True):
   # Extract the href (link) from the current anchor tag
    href = a["href"]


    # Check whether the link corresponds to a movie page
    # Movie URLs on JustWatch contain '/movie/'
    if "/movie/" in href:

        # Convert the relative URL into a complete (absolute) URL
        full_url = "https://www.justwatch.com" + href

        # Add the full movie URL to the list
        movie_urls.append(full_url)

# Remove duplicate movie URLs by converting the list to a set and back to a list
movie_urls = list(set(movie_urls))

# Display the total number of unique movie URLs extracted
print(len(movie_urls))

# Create a final list containing all the extracted movie URLs
# (This step is optional since movie_urls already contains the required data)
final_url = [url for url in movie_urls]

# Print the list of all extracted movie URLs
print(final_url)
```

## **Scrapping IMBD Rating**

```python
# Create an empty list to store the IMDb ratings of all movies
imdb_ratings = []

for movie_url in movie_urls:

    response = requests.get(movie_url, headers=headers)

    soup = BeautifulSoup(response.text, "html.parser")

    # Find the IMDb logo image (<img> tag with alt="IMDB")
    # This helps locate the IMDb rating section on the page

    imdb_tag = soup.find("img", alt="IMDB")

    # Check whether the IMDb section exists
    if imdb_tag:

        # Find the next <div> element that contains the IMDb rating            rating_div = imdb_tag.find_next("div")

        rating_div = imdb_tag.find_next("div")

        # Extract the IMDb rating text, remove extra spaces,
        # and keep only the rating value
        # Example: "7.8 (171k)" -> "7.8"

        imdb_rating = rating_div.get_text(strip=True).split()[0]

        # Add the extracted IMDb rating to the list
        imdb_ratings.append(imdb_rating)

    else:
        # If the IMDb rating is not available, store None
        imdb_ratings.append(None)

# Print the list of all extracted IMDb ratings
print(imdb_ratings)


 # Print the total number of IMDb ratings extracted
print(len(imdb_ratings))
```

## **Scrapping Age Rating**

```python
# Create an empty list to store the age ratings of all movies
age_ratings = []

for movie_url in movie_urls:

    response = requests.get(
        movie_url,
        headers=headers
    )

    soup = BeautifulSoup(
        response.text,
        "html.parser"
    )
    # Find the <h3> tag whose text is "Age rating"
    # This acts as the heading for the age rating section

    age_heading = soup.find(
        "h3",
        class_="poster-detail-infos__subheading",
        string="Age rating"
    )
  # Check whether the Age Rating section exists
    if age_heading:
        # Find the next <div> that contains the age rating value
        age_div = age_heading.find_next(
            "div",
            class_="poster-detail-infos__value"
        )
       # Extract the age rating text (e.g., "UA16+", "A", "PG-13")
        # and add it to the age_ratings list
        age_ratings.append(
            age_div.get_text(strip=True)
        )

    else:
        # If the age rating is not available, store None
        age_ratings.append(None)

# Print the total number of age ratings extracted
print(len(age_ratings))

# Print the age rating of each movie
for age_rating in age_ratings:
  print(age_rating)
```

## **Fetching Production Countries Details**

```python
# Create an empty list to store the production country of all movies
production_countries = []

for movie_url in movie_urls:

    response = requests.get(
        movie_url,
        headers=headers
    )

    soup = BeautifulSoup(
        response.text,
        "html.parser"
    )

    # Find the <h3> tag whose text is "Production country"
    # This acts as the heading for the production country section
    country_heading = soup.find(
        "h3",
        class_="poster-detail-infos__subheading",
        string="Production country"
    )

    # Check whether the Production Country section exists
    if country_heading:
        # Find the next <div> that contains the production country information
        country_div = country_heading.find_next(
            "div",
            class_="poster-detail-infos__value"
        )

        # Extract the production country text
        # (e.g., "United States", "India", "United Kingdom")
        # and add it to the production_countries list
        production_countries.append(
            country_div.get_text(strip=True)
        )

    else:
      # If the production country is not available, store None
        production_countries.append(None)

# Print the total number of production country entries extracted
print(len(production_countries))


# Print the production country of each movie
for production_country in production_countries:
    print(production_country)
```

## **Fetching Streaming Service Details**

```python
# Create an empty list to store the streaming services available for each movie
streaming_services = []

# Loop through each movie URL extracted from the JustWatch website
for movie_url in movie_urls:

    # Send an HTTP GET request to the current movie page
    response = requests.get(
        movie_url,
        headers=headers
    )

    # Parse the HTML content of the movie page
    soup = BeautifulSoup(
        response.text,
        "html.parser"
    )

    # Find all provider icons (<img> tags) that represent
    # the streaming platforms where the movie is available
    provider_imgs = soup.find_all(
        "img",
        class_="provider-icon"
    )

    # Extract the value of the 'alt' attribute from each image
    # The alt attribute contains the name of the streaming service
    providers = [
        img.get("alt")
        for img in provider_imgs
    ]

    # Combine all provider names into a single comma-separated string
    # and store it in the streaming_services list
    streaming_services.append(
        ", ".join(providers)
    )

# Define a set of valid OTT/streaming platforms
# Only these platforms will be considered while selecting the primary OTT service
valid_services = {
    "Netflix",
    "Amazon Prime Video",
    "Amazon Prime Video with Ads",
    "JioHotstar",
    "Zee5",
    "Sun Nxt",
    "Sony LIV",
    "Aha",
    "Apple TV Store",
    "Amazon Video",
    "Hoichoi",
    "Eros Now",
    "MX Player",
    "Plex Player",
    "JustWatch",
    "Crunchyroll",
}

# Create an empty list to store the primary OTT platform for each movie
primary_ott = []

# Loop through the extracted streaming service strings
for services in streaming_services:

    # Split the comma-separated string into a list of individual services
    service_list = [
        s.strip()
        for s in services.split(",")
    ]

    # Find the first service that matches one of the valid OTT platforms
    # If no valid service is found, return None
    ott = next(
        (
            s
            for s in service_list
            if s in valid_services
        ),
        None
    )

    # Store the selected primary OTT platform
    primary_ott.append(ott)

# Print the total number of primary OTT entries extracted
print("Total Primary OTT Platforms Identified:", len(primary_ott))

# Print the primary OTT platform for each movie
print("Primary OTT Platform for Each Movie:")
for main_ott in primary_ott:
    print(main_ott)
```
