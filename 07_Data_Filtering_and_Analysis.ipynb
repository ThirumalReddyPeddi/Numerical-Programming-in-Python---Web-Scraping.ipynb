## **Task 2 :- Data Filtering & Analysis**

```python
# Import the required libraries
# Pandas is used for data manipulation and analysis.
# datetime is used to work with date and time values.

import pandas as pd
from datetime import datetime

# Convert the 'Release_Year' column in the Movies DataFrame
# from text (string) format to numeric format.
# Any invalid or missing values will be converted to NaN.

movies_df["Release_Year"] = pd.to_numeric(
    movies_df["Release_Year"],
    errors="coerce"
)

# Convert the 'IMDb_Rating' column in the Movies DataFrame
# from text (string) format to numeric format.
# Invalid or missing values are converted to NaN.

movies_df["IMDb_Rating"] = pd.to_numeric(
    movies_df["IMDb_Rating"],
    errors="coerce"
)

# Convert the 'Release_Year' column in the TV Shows DataFrame
# from text (string) format to numeric format.
# Invalid or missing values are converted to NaN.

tv_shows_df["Release_Year"] = pd.to_numeric(
    tv_shows_df["Release_Year"],
    errors="coerce"
)

# Convert the 'IMDb_Rating' column in the TV Shows DataFrame
# from text (string) format to numeric format.
# Invalid or missing values are converted to NaN.

tv_shows_df["IMDb_Rating"] = pd.to_numeric(
    tv_shows_df["IMDb_Rating"],
    errors="coerce"
)
```

```python
# Step 2: Get the current year and calculate the cutoff year

current_year = datetime.now().year
cutoff_year = current_year - 2
```

```python
# Step 3: Filter Movies

filtered_movies_df = movies_df[
    (movies_df["Release_Year"] >= cutoff_year) &
    (movies_df["IMDb_Rating"] >= 7)
]

filtered_movies_df.head()
```

```python
# Step 4: Filter TV Shows

filtered_tv_shows_df = tv_shows_df[
    (tv_shows_df["Release_Year"] >= cutoff_year) &
    (tv_shows_df["IMDb_Rating"] >= 7)
]

filtered_tv_shows_df.head()
```

## **Calculating Mean IMDB Ratings for both Movies and Tv Shows**

```python
movies_mean = movies_df["IMDb_Rating"].mean()
tv_shows_mean = tv_shows_df["IMDb_Rating"].mean()

print("Mean IMDb Rating for Movies:", round(movies_mean, 2))
print("Mean IMDb Rating for TV Shows:", round(tv_shows_mean, 2))
```

## **Analyzing Top Genres**

```python
# Top Genres for Movies

movie_genres = (
    movies_df["Genres"]
    .dropna()
    .str.split(", ")
    .explode()
)

# Count the occurrence of each genre in the dataset.
# The value_counts() function returns the frequency of each unique genre,
# sorted in descending order (most common genre first).

top_movie_genres = movie_genres.value_counts()

print(top_movie_genres)
```

```python
# Extract and analyze the genres present in the TV Shows DataFrame.

# Select the 'Genres' column from the DataFrame.
# Remove any missing (NaN) values to avoid errors during processing.
# Split the comma-separated genre strings into individual genres.
# Example:
# "Drama, Mystery, Thriller"
# becomes
# ["Drama", "Mystery", "Thriller"]
# The explode() function then creates a separate row for each genre.
tv_genres = (
    tv_shows_df["Genres"]
    .dropna()
    .str.split(", ")
    .explode()
)


top_tv_genres = tv_genres.value_counts()

# Display the frequency count of all TV show genres.
print("Top TV Show Genres:")
print(top_tv_genres)
```

```python
# Top Genres for TV Shows
tv_genres = (
    tv_shows_df["Genres"]
    .dropna()
    .str.split(", ")
    .explode()
)
# Count the occurrence of each genre in the dataset.
# The value_counts() function returns the frequency of each unique genre,
# sorted in descending order (most common genre first).

top_tv_genres = tv_genres.value_counts()

print(top_tv_genres)
```

```python
# Top 5 Genres
print("Top 5 Movie Genres:")

# Display the five most frequently occurring genres in the Movies dataset.
print(top_movie_genres.head(5))

print("\nTop 5 TV Show Genres:")

# Display the five most frequently occurring genres in the TV Shows dataset.
print(top_tv_genres.head(5))
```

```python
movie_streaming_counts = (
    movies_df['OTT Streaming_Service']
    .dropna()
    .str.split(", ")
    .explode()
    .value_counts()
)

print(movie_streaming_counts)
```

```python
# Most Predominant Streaming Service:

print("Predominant Streaming Service for Movies:")

# The idxmax() method returns the index label corresponding
# to the maximum value in the Series.
# In this case, it returns the OTT platform with the highest
# number of movie titles.
print(movie_streaming_counts.idxmax())


print("Number of Titles:", movie_streaming_counts.max())
```

```python
# For TV Shows
tv_streaming_counts = (
    tv_shows_df['OTT_Streaming_Service']
    .dropna()
    .str.split(", ")
    .explode()
    .value_counts()
)

print(tv_streaming_counts)
```

```python
# Most Predominant Streaming Service

print("Predominant Streaming Service for TV Shows:")
print(tv_streaming_counts.idxmax())
print("Number of Titles:", tv_streaming_counts.max())
```
