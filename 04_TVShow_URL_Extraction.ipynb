## **Scraping TV  Show Data**

```python
!pip install beautifulsoup4
!pip install requests
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np


# URL from which tv shows data is fetched
tv_url='https://www.justwatch.com/in/tv-shows?release_year_from=2000'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
}
# Sending an HTTP GET request to the URL
response = requests.get(tv_url,headers=headers)

# Parsing HTML content with Beautiful Soup
soup_tv = BeautifulSoup(response.text,'html.parser')
print(soup_tv.prettify())
```

```python
# Display the HTTP response status code returned by the server
# A status code of 200 indicates that the request was successful.
# Other common status codes include:
# 404 - Page Not Found
# 403 - Access Forbidden
# 500 - Internal Server Error

print(response.status_code)

# Display the final URL from which the response was received.
# This is useful for checking whether the request was redirected
# to a different page before the content was fetched.

print(response.url)
```

## **Fetching Tv shows Url details**

```python
# Create an empty list to store the URLs of all TV show pages
tv_show_urls = []


# Find all anchor (<a>) tags that contain an href attribute
# and iterate through each link on the webpage

for link in soup_tv.find_all('a', href=True):
   # Extract the value of the href attribute
    href = link.get('href')


    # Check whether the link corresponds to a TV show page
    # TV show URLs on JustWatch contain '/tv-show/'

    if href and '/tv-show/' in href:

        # Convert the relative URL into a complete (absolute) URL
        # by appending the JustWatch domain name

        tv_show_urls.append('https://www.justwatch.com' + href)


# Remove duplicate TV show URLs by converting the list
# to a set and then back to a list

tv_show_urls = list(set(tv_show_urls))

# Print the total number of unique TV show URLs extracted
print(len(tv_show_urls))

# Print the list of all extracted TV show URLs
print(tv_show_urls)
```
